```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import yfinance as yf
```

```python
stock = yf.Ticker("0050.TW")
data50 = stock.history(period="max")
data50['Adj Close'] = data50['Close'] * 1.6666
data50 = data50[['Open', 'Close', 'Volume', 'Adj Close']]

stock = yf.Ticker("0053.TW")
data53 = stock.history(period="max")
data53['Adj Close'] = data53['Close'] * 3.826
data53 = data53[['Open', 'Close', 'Volume', 'Adj Close']]


stock = yf.Ticker("0055.TW")
data55 = stock.history(period="max")
data55['Adj Close'] = data55['Close'] * 7.8125
data55 = data55[['Open', 'Close', 'Volume', 'Adj Close']]

stock = yf.Ticker("0051.TW")
data51 = stock.history(period="max")
data51['Adj Close'] = data51['Close'] * 3.0176
data51 = data51[['Open', 'Close', 'Volume', 'Adj Close']]

stock = yf.Ticker("^TWII")
dataIndex = stock.history(period="max")
dataIndex = dataIndex['2008-01-02':]
dataIndex['Adj Close'] = dataIndex['Close'] / 83.23
dataIndex = dataIndex[['Open', 'Close', 'Volume', 'Adj Close']]


plt.plot(dataIndex['Adj Close'], label = 'TW Index')
plt.plot(data50['Adj Close'], label = '0050.TW')
plt.plot(data53['Adj Close'], label = '0053.TW')
plt.plot(data55['Adj Close'], label = '0055.TW')
plt.plot(data51['Adj Close'], label = '0051.TW')


plt.legend()

plt.show()
```
